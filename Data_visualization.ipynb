```python
x1=[1,2,3,4]
y1=[10,20,40,60]
plt.plot(x1, y1,label= 'First Line')
x2=[4,6,7,9]
y2=[20,30,50,80]
plt.plot(x2, y2,label= 'Second Line')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot the Graph')
plt.legend()   
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]

x2 = [2, 4, 6, 8]
y2 = [1, 2, 3, 4]
plt.plot(x1, y1, label = 'First line')  # → label: assigns name to the line

plt.plot(x2, y2, label = 'Second Line')
plt.xlabel('X-axis')                    # → xlabel: Assign label to x - axis
plt.ylabel('Y-Axis')                    # → ylabel: Assign label to y - axis
plt.title('Simple Graph')               # → title: The plot’s title

plt.legend()                            # → legend: Places a legend in the axes
plt.show()
```

```python
x= [2,3,4,5,6]
y= [4,9,16,25,36]

plt.plot(x, y, color= 'k', ls= '--', lw= 2, marker= 'o', markersize= 8)  # Adding marker and markersize

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot the Graph')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D','E']

values= [20,25,10,5,40]

plt.bar(categories, values,color='red')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar Graph')

plt.show()
```

```python
# Creating basic Scatter Plot using lists

x = [2,4,6,8,10]  
y = [4,20,30,5,45]  
x2 = [2, 4, 6, 8]
y2 = [6, 30, 50, 43]
plt.scatter(x, y, color= 'b')
plt.scatter(x2, y2, color= 'r')
plt.axis([0,25,0,65])
  
plt.title('Scatter Plot')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')    
plt.show()
```

```python
x = [2,4,6,8,10]  
y = [4,20,30,5,45]  
  
x2 = [5,10,12,15,20]  
y2 = [20,2,36,60,42]  
  
plt.scatter(x, y, color= 'r') 
plt.axis([0,25,0,65])
  
plt.scatter(x2, y2, color='b')  
  
plt.title('Scatter Plot comparing two Datasets')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')  
  
plt.show()
```

```python
time_spent = [240,360,50,60,60]

activities = ['Social_Media', 'TV', 'Education', 'Sports', 'Food']

plt.pie(time_spent, labels= activities)

plt.title('Students Activity Data')

plt.show()
```

```python
marks=[90,45,23,12,5]
grades=['Dist','Idiv','II-div','pass','fail']
plt.pie(marks,autopct='%.2f%%',labels=grades,explode=[0,0.1,0,0,0])
plt.title('Students Obtained Marks')

plt.show()
```

```python
import matplotlib.pyplot as plt
ecommerce=['maytra', 'Amazon', 'snapdeal', 'alibaba', 'flipcart']
Q1_profit =[35, 45, 100, 70, 40]
Q2_profit =[38, 40, 105, 65, 45]

Q3_profit =[30, 42, 120, 72, 50]
Q4_profit =[25, 34, 115, 60, 48]
plt.figure(1, figsize=(10,10))
plt.subplot(221)
plt.bar(ecommerce, Q1_profit)
plt.title('Quarter1  profit')

plt.subplot(222)
plt.bar(ecommerce, Q2_profit)
plt.title('Quarter2  profit')


plt.subplot(223)
plt.bar(ecommerce, Q3_profit)
plt.title('Quarter3  profit')

plt.subplot(224)
plt.bar(ecommerce, Q4_profit)
plt.title('Quarter4  profit')

```
