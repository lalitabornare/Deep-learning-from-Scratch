```python
#Experiment 1 : To plot different activation function used in ML/DL
import numpy as np
import matplotlib.pyplot as plt
#1. Binary Sigmoidal Activation
plt.subplot(3,3,1)
#plt.figure(figsize=(8,4))
def SigmoidBinary(t):
  return 1/(1+np.exp(-t))
t=np.linspace(-10,10,100)

plt.plot(t,SigmoidBinary(t),label='Sigmoid')
plt.title('Binary Sigmoid Activation Function')
plt.style.use('seaborn-v0_8')

plt.show()

#2. Hyperbolic Tangent Activation
plt.subplot(3,3,2)
#plt.figure(figsize=(8,4))
def HyperbolicTan(t):
  return (np.exp(t)-np.exp(-t))/(np.exp(t)+np.exp(-t))
t=np.linspace(-10,10,100)

plt.plot(t,HyperbolicTan(t),label='Hyperbolic')
plt.title('Hyperbolic Tangent Activation Function')
plt.style.use('seaborn-v0_8')

plt.show()

#3.LeakyRelu Activation
plt.subplot(3,3,3)
#plt.figure(figsize=(8,4))
def LeakyRectifiedLinearUnit(arr):
  lst=[]
  for i in arr:
    if i>=0:
      lst.append(i)
    else:
      lst.append(0.1*i)
  return lst
arr=np.linspace(-10,10,100)
plt.plot(arr,LeakyRectifiedLinearUnit(arr),label='LeakyRectifiedLinearUnit')
plt.title('LeakyRelu Activation Function')
plt.style.use('seaborn-v0_8')
plt.show()

#4.Relu Activation
plt.subplot(3,3,4)
#plt.figure(figsize=(8,4))
def ReLU(arr):
  return np.maximum(0,arr)
#Derivative for ReLU
def der_ReLU(x):
  data = [1 if value>0 else 0 for value in x]
  return np.array(data, dtype=float)
arr=np.linspace(-10,10,100)
dy_data = der_ReLU(arr)
plt.plot(arr, dy_data)
plt.plot(arr,ReLU(arr),label='RELU')
plt.title('Relu Activation Function & Derivative')
plt.style.use('seaborn-v0_8')
plt.show()

#5.Softmax Activation
plt.subplot(3,3,5)
#plt.figure(figsize=(8,4))
def softmax(t):
  return np.exp(t)/np.sum(np.exp(t))
t=np.linspace(-10,10,100)
plt.plot(t,softmax(t),label='Softmax')
plt.title('Softmax Activation Function')
plt.style.use('seaborn-v0_8')
plt.show()

```
